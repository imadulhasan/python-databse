```python
fig = plt.figure(figsize=(8,3))
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y, linestyle='--', color='blue', linewidth=2, label='sin(x)')
plt.title('time series')
plt.xlabel('time')
plt.ylabel('value')
plt.legend()
```

```python


fig = plt.figure(figsize=(8,3))
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,z, linestyle=':', color='red', linewidth=1, label='cos(x)')
plt.plot(x,y, linestyle='--', color='blue', linewidth=2, label='sin(x)')
plt.title('time series' , loc='left', pad=10, fontdict={'fontsize':16,
                                                        'color':'blue',
                                                        'weight':'bold'})
plt.xlabel('time')
plt.ylabel('value')
plt.legend()
```

```python
fig = plt.figure(figsize=(8,3))
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,z, linestyle=':', color='red', linewidth=1, label='cos(x)')
plt.plot(x,y, linestyle='--', color='blue', linewidth=2, label='sin(x)')
plt.xlim(-1,12)
plt.ylim(-1.5,1.5)
plt.legend()
```

```python
rainfall = [0.8, 0.6, 0,0,0.1,0.2,1.2,1.5,0.1,0.3,0.6,0.8]
months=['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
fig=plt.figure(figsize=(8,3))
plt.bar(months,rainfall,color='blue',hatch='//')
plt.title('rainfall in title')
plt.xlabel('month')
plt.ylabel('rainfall(mm)')
```
