```python
df = pd.read_json('flipkart_fashion_products_dataset.json')
df


```

```python
#1 - removing unwanted columns
cols_to_delete = ['_id', 'pid', 'url', 'images']
df.drop(cols_to_delete, axis=1, inplace=True)
df

```

```python
def clean_price(price):
    if price is None:
        return np.nan
    elif isinstance(price, str):
        price = price.replace(',','')
        if price:
            price = float(price)
        else:
            price = np.nan
        return price

df['actual_price'] = df['actual_price'].apply(clean_price)            
```

```python
def clean_discount(price):
    if price is None:
        return np.nan
    elif isinstance(price, str):
        price = price.replace('% off','')
        if price:
            price = float(price)
        else:
            price = np.nan
        return price
df ['discount'] = df['discount'].apply(clean_discount) 
```

```python
def clean_price(selling):
    if selling is None:
        return np.nan
    elif isinstance(selling , str):
        selling = selling.replace(',','')
        if selling:
            selling = float(selling)
        else:
            selling = np.nan
        return selling
    
df ['selling_price'] = df['selling_price'].apply(clean_price)      
```

```python
def clean_rating(price):
    if price is None:
        return np.nan
    elif isinstance(price, str):
        price = price.replace(',','')
        if price:
            price = float(price)
        else:
            price = np.nan
        return price

df ['average_rating'] = df['average_rating'].apply(clean_rating)
```

```python
pricesdf[['actual_price','selling_price']].plot.kde()
plt.xlim(5000, 25000)
plt.xlabel('actual_price')
plt.text(15000,0.00015, 'Distribution of\nactual price',fontsize=20,fontweight='bold',color='orange')
conclusion= 'the distribution is right skewed'
plt.text(15000,0.00012,conclusion,fontsize=9)
plt.show()
```
